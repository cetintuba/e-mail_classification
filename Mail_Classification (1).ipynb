# E-Mail classification NLP project

```python
# Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score, classification_report
import re
```

```python
# read data
train_df = pd.read_csv('SMS_train.csv', encoding='latin-1')
test_df = pd.read_csv('SMS_test.csv', encoding='latin-1')
```

```python
train_df.head()
```

```python
# replace
train_df['Message_body'] = train_df['Message_body'].str.lower().replace('[^\w\s]', '', regex=True).replace('\d+', '', regex=True).replace('\n', '', regex=True).replace('\r', '', regex=True)
train_df.head(5)
```

```python
# encoding
train_df['Label'] = train_df['Label'].map({'Non-Spam': 0, 'Spam': 1})
```

```python
# # Check missing values
train_df.isnull().sum()
```

```python
# x y split
x = train_df["Message_body"]
y = train_df["Label"]
```

```python
# vectorizing
vect=CountVectorizer(stop_words="english")
x=vect.fit_transform(x)
```

```python
# train test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
# model
from sklearn.svm import SVC
s=SVC()
model = s.fit(x_train, y_train)
```

```python
#make predictions
predictions=model.predict(x_test)
```

```python
# model accuracy
accuracy_score(predictions,y_test)
```

```python
#save model
import joblib
joblib.dump(model,"model.joblib")
```
